```python
import pandas as pd

url = "https://www.census.gov/foreign-trade/balance/c0015.html"
tables = pd.read_html(url)
len(tables)  # número de tablas detectadas
tables[0].head()  # muestra la primera tabla

df = pd.concat(tables, ignore_index=True, sort=False)

df
```

```python
df = df[~df['Month'].str.startswith('TOTAL')]
df
```

```python
# exportar a csv
df.to_csv('TB_USA.csv', index=False)
```
